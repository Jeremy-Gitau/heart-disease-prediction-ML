```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("heart.csv")
```

```python
data.head(5)
```

```python
data.tail(5)
```

```python
le = LabelEncoder()
# labels = le.fit_transform(data['AHD'])
chestpain = le.fit_transform(data['ChestPain'])
le.classes_
```

```python
thal = le.fit_transform(data['Thal'])
le.classes_
```

```python
# data['AHD'] = labels
data['ChestPain'] = chestpain
data['Thal'] =thal
data = data.drop(columns = 'Ca', axis = 1)
data = data.drop(columns = 'Unnamed: 0', axis = 1)
data
```

```python
data['AHD'].value_counts()
```

```python

new = str(data['Sex']).str.split("1",  n = 1, expand = True)
new
```

```python
X = data.drop(columns="AHD", axis=1)
Y = data['AHD']
```

```python
X.head(2)
```

```python
Y.head(3)
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y, test_size=0.6, stratify=Y,random_state=2)
```

```python
model = LogisticRegression()
model.fit(X_train,Y_train)
prediction = model.predict(X_test)
accuracy = accuracy_score(prediction,Y_test)
accuracy
```

```python
input_data =(63,1,3,145,233,1,2,150,0,2.3,3,0)
input_data_array = np.asarray(input_data)
reshaped_input = input_data_array.reshape(1,-1)
```

```python
pred = model.predict(reshaped_input)
if (pred[0] == "Yes"):
    print("you have a heart disease")
else:
        print("you are healthy")
```
